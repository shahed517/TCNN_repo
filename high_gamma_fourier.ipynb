```python
import numpy as np
path = f'/scratch/gilbreth/ahmed348/Dutch_dataset_features/TCNN_folder/FEATURES/features_HGA_ONLY/sub-05_feat.npy'
signal = np.load(path)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assume `signal` has shape (30030, 127)
# signal = ...

fs = 100  # sampling frequency in Hz
T, C = signal.shape

# Randomly select a channel
ch_idx = np.random.randint(C)
print(ch_idx)
x = signal[:, ch_idx]

# FFT
X = np.fft.fft(x)
freqs = np.fft.fftfreq(T, d=1/fs)

# Keep only positive frequencies
pos_mask = freqs >= 0
freqs_pos = freqs[pos_mask]
X_mag = np.abs(X[pos_mask]) / T  # normalized magnitude

# Plot
plt.figure(figsize=(7, 4))
plt.plot(freqs_pos, X_mag)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title(f"Fourier Spectrum (Channel {ch_idx})")
plt.grid(True)
plt.tight_layout()
plt.show()
```
